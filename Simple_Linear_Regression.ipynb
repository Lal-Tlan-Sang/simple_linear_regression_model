```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
X = data_mart.drop(["ProductID","OutletSales"],axis =1)
Y =data_mart["OutletSales"]
```

```python
X = sm.add_constant(X)
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,train_size =0.7,random_state =12)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
model = sm.OLS(y_train,X_train[['MRP','const']])
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
results.predict(X_train[['MRP','const']])
```

```python
results.predict(X_test[['MRP','const']])
```

```python
def mod_metrics(act,pred):
  print("RMSE:",np.sqrt(((act-pred)**2).mean()),"\n")
  print("MAE:",(np.abs(act-pred).mean()),"\n")
  print("MAPE:",(np.abs((act-pred)/act).mean()),"\n")
```

```python
mod_metrics(y_train,results.predict(X_train[['MRP','const']]))
```

```python
mod_metrics(y_test,results.predict(X_test[["MRP",'const']]))
```
